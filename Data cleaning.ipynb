```python
import pandas as pd 
import numpy as np
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv('netflix1.csv')
df
```

```python
df.loc[df['director'] == 'Not Given'].count()
```

```python
df['director'] = df['director'].replace({'Not Given' :  'Unknow'})
df
```

```python
df = df.drop(columns=['director', 'rating'])
df
```

```python
df.info()
```

```python
df['release_year'] = pd.to_datetime(df['release_year'],format = '%Y')
df['date_added'] = pd.to_datetime(df['date_added'],format = '%m/%d/%Y')
```

```python
""" df['release_year'] = df['release_year'].dt.strftime("%Y") """
df
```

```python
df['genre_list'] = df['listed_in'].str.split(',').str[0]
df
```

```python
df = df.drop(columns='listed_in')
df
```

# Eda

### Distribución de los Géneros

```python
df_genre_distribution = df.groupby('genre_list').count()
df_genre_distribution = df_genre_distribution.rename(columns={"show_id" : 'Number Of Movies'})
df_genre_distribution = df_genre_distribution[['Number Of Movies']].sort_values(by='Number Of Movies' ,ascending=False)
df_genre_distribution = df_genre_distribution.reset_index()
df_genre_distribution.head()
```

```python
plt.bar(df_genre_distribution["genre_list"], df_genre_distribution["Number Of Movies"], width=0.6, color="#E50914")
plt.title("Count of Movies")
plt.xticks(df_genre_distribution['genre_list'], rotation=90)
plt.xlabel("genre_list")
plt.ylabel("Number Of Movies")   
```

### Duración de Películas y Series

```python
df_duration_Movies = df.loc[df['type']=='Movie']
df_duration_Movies['duration'] = df['duration'].str.split(' ').str[0]
df_duration_Movies['duration'] = df_duration_Movies['duration'].astype('int64')
df_duration_Movies = df_duration_Movies[['type', 'title', 'duration']]
df_duration_Movies.head()
```

```python
df_duration_Serie = df.loc[df['type']=='TV Show']
df_duration_Serie['duration'] = df['duration'].str.split(' ').str[0]
df_duration_Serie['duration'] = df_duration_Serie['duration'].astype('int64')
df_duration_Serie = df_duration_Serie[['type', 'title', 'duration']]
df_duration_Serie 		
```

```python
df_duration_Serie.describe()
```

### Lanzamientos por País

```python
df_country = df.groupby('country').count()
df_country = df_country.rename(columns={"show_id" : 'Number Of Movies'})
df_country = df_country[['Number Of Movies']].sort_values(by='Number Of Movies' ,ascending=False)
df_country = df_country.reset_index()
df_country
```

### Tendencias de Fechas de Adición

```python
df_date_adition = df.groupby('date_added').count()
df_date_adition = df_date_adition.rename(columns={"show_id" : 'Number of adition'})
df_date_adition = df_date_adition[['Number of adition']].sort_values(by='Number of adition' ,ascending=False)
df_date_adition = df_date_adition.reset_index()
df_date_adition 
```

### Comparación de Años de Lanzamiento

```python
df_release = df.groupby('release_year').count()
df_release = df_release.rename(columns={"show_id" : 'Count of release'})
df_release = df_release[['Count of release']].sort_values(by='Count of release' ,ascending=False)
df_release = df_release.reset_index()
df_release 
```

```python
plt.plot(df_release['release_year'], df_release['Count of release'], color="red")
plt.title("Number of movies according to the years")
plt.xlabel("release_year")
plt.ylabel("Number of adition")
```

#### Exportar


```python
dataframes = [df, df_date_adition, df_country, df_duration_Movies, df_duration_Serie, df_genre_distribution , df_release]
with pd.ExcelWriter('df_clean.xlsx') as writer:
    for i, df in enumerate(dataframes):
        sheet_name = f'Sheet{i+1}'  # Nombre de la hoja
        df.to_excel(writer, sheet_name=sheet_name, index=False)
```
